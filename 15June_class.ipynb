```python
import pandas as pd

df1 = pd.DataFrame({
    'id' : [1,2,3,4],
    'name' : ['Raja', 'Rani','Rocky', 'Rahul'],
    'age': [25,30, 35,40]
})


df2 = pd.DataFrame({
    'id':[3,4,5,6],
    'name': ['Rahul', 'Reena', 'Raja', 'Rocky'],
    'age' : [35,40,45,50]
})

df3 = pd.merge(df1, df2, on = 'id', how = 'inner')  #inner join

```

```python
df3.rename(columns={'name_x': 'table_name1'}, inplace=True)
```

```python
df4 = pd.merge(df1, df2, on = 'id', how = 'left')
print(df4)
```

```python
df5 = pd.merge(df1, df2, on = 'id', how = 'right')
print(df5)
```

```python
df6 = pd.merge(df1, df2, on = 'id', how = 'outer')
print(df6)
```

```python
df7 = pd.merge(df1, df2, how = 'cross')
display(df7)
```
