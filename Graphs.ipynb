```python
batch1 = [10, 20, 30, 40, 50]
batch2 = [15, 25, 35, 45, 55]

box = plt.boxplot(
    [batch1, batch2],
    tick_labels=["Batch 1", "Batch 2"],
    patch_artist=True
)

colors = ['#A8DADC', '#CDB4DB']

for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

for median in box['medians']:
    median.set_color('black')
    median.set_linewidth(2)

plt.title("Batch Comparison", fontsize=14)
plt.grid(alpha=0.2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
data = np.random.rand(10, 10)
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
im1 = axes[0].imshow(data, cmap='gray')
axes[0].set_title("Grayscale ('gray')")
fig.colorbar(im1, ax=axes[0])
im2 = axes[1].imshow(data, cmap='coolwarm')
axes[1].set_title("Cmap ('coolwarm')")
fig.colorbar(im2, ax=axes[1])
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv("student_data.csv")
df.head()
```

```python
sns.lineplot(data = df, x = 'study_hours', y = 'exam_score')
plt.title('Line Plot of Study Hours vs Exam Score')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.show()
```

```python
def grade(score):
  if score >= 90:
    return 'A'
  elif score >= 80:
    return 'B'
  elif score >= 70:
    return 'C'
  elif score >= 60:
    return 'D'
  else:
    return 'F'

df['grade'] = df['exam_score'].apply(grade)
# Bar plot

sns.barplot(data = df, x = 'grade', y = 'exam_score', estimator='mean')

plt.title('Performance')

plt.tight_layout()
plt.show()
```
