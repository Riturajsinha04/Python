```python
a = np.random.randint(5, 15, 6)
b = np.random.randint(5, 15, 6)
c = np.concatenate((a, b), axis=0)
print(a)
print(b)
print(c)
```

```python
a = np.random.randint(5, 15, 6).reshape(2, 3)
b = np.random.randint(5, 15, 6).reshape(2, 3)
c = np.concatenate((a, b), axis=1)
print(a)
print(b)
print(c)
```

```python
a = np.random.randint(2, 15, 6)
b = np.random.randint(1, 19, 6)
c = np.where(b > 2, b, a)
print(a)
print(b)
print(c)
```

```python
scores = []

days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]

for day in days:
    score = int(input(f"Enter the score for {day}: "))
    scores.append(score)
    score = np.array(scores)

    print("Scores", scores)
    print("Maximum Score", np.max(score))
    print("Minimum Score", np.min(score))
    print("Average Score", np.mean(score))
    print("Standard deviation", np.std(score))
```
